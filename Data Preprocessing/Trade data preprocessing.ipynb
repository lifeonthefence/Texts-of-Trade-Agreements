```python
import pandas as pd

# import from gravity model dataset
data = pd.read_csv('../gravity/gravity.csv', low_memory=False)
```

```python
# use trade agreement data to filter out irrelevant pairs
d = pd.read_csv('fta_clean.csv')

pairs = list(d.apply(lambda r: (r['Party 1'], r['Party 2']), axis=1))

data = data[data.apply(lambda r: (r['iso3_o'], r['iso3_d']), axis=1).isin(pairs)]
data = data.reset_index(drop=True)

data.to_csv('trade_clean.csv')
```

```python
# list pairs of data absent
pairs_with_data = list(data.groupby(['iso3_o', 'iso3_d']).groups.keys())
set(pairs) - set(pairs_with_data)
```

```python
# print selected data for inspection
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    print(data[['year', 'iso3_o', 'iso3_d', 'tradeflow_comtrade_o', 'tradeflow_comtrade_d', 'tradeflow_imf_o', 'tradeflow_imf_d']])
```
